```python
categorical_columns_with_missing = ['Parental_Education_Level', 'Distance_from_Home', 'Teacher_Quality']

for column in categorical_columns_with_missing:
    mode_value = df[column].mode()[0]  
    df.fillna({column: mode_value}, inplace=True)
df.sample(5)

#Bu kod, belirtilen kategorik sÃ¼tunlardaki eksik deÄŸerleri, her sÃ¼tunun en sÄ±k gÃ¶rÃ¼len deÄŸeri (mod) ile doldurur ve ardÄ±ndan veri Ã§erÃ§evesinin gÃ¼ncellenmiÅŸ halinden 5 rastgele satÄ±r Ã¶rneÄŸi gÃ¶sterir. 
```

```python
categorical_columns = [
    'Parental_Involvement', 'Access_to_Resources', 'Extracurricular_Activities', 
    'Motivation_Level', 'Internet_Access', 'Family_Income','Teacher_Quality', 'School_Type', 
    'Peer_Influence', 'Learning_Disabilities', 'Parental_Education_Level','Distance_from_Home','Gender'
]

df = pd.get_dummies(df, columns=categorical_columns, drop_first=True)
df.sample(5)

#Bu kod, kategorik deÄŸerleri One-Hot Encoding yÃ¶ntemiyle dÃ¶nÃ¼ÅŸtÃ¼rÃ¼r.Bu iÅŸlem, kategorik sÃ¼tunlardaki her bir benzersiz deÄŸeri ayrÄ± bir sÃ¼tun olarak ele alÄ±r ve bu sÃ¼tunlara 1 ve 0 deÄŸerlerini atar.
```

```python
z = np.abs((df - df.mean()) / df.std())
threshold = 3
df= df[(z < threshold).all(axis=1)]

#Bu kod, veri setindeki her bir sayÄ±sal sÃ¼tun iÃ§in Z-score deÄŸerlerini hesaplar ve bu deÄŸerlerin belirli bir eÅŸik deÄŸerinden bÃ¼yÃ¼k olup olmadÄ±ÄŸÄ±nÄ± kontrol eder. EÅŸik deÄŸeri genellikle 3 olarak belirlenir. EÄŸer bir veri noktasÄ±nÄ±n Z-score deÄŸeri 3'ten bÃ¼yÃ¼kse, bu veri noktasÄ± uÃ§ deÄŸer olarak kabul edilir ve veri setinden Ã§Ä±karÄ±lÄ±r.
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, accuracy_score
from sklearn.metrics import f1_score, precision_score, recall_score
```

```python
X = df.drop("Exam_Score",axis=1)
y = df[["Exam_Score"]]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
y_train = y_train.squeeze()
y_test = y_test.squeeze()

#EÄŸer y_train bir numpy array ve ÅŸekli (n, 1) (n Ã¶rnek, 1 hedef deÄŸiÅŸken) ise, y_train.squeeze() bu diziyi (n,) ÅŸeklinde dÃ¶nÃ¼ÅŸtÃ¼rÃ¼r. AynÄ±sÄ± y_test iÃ§in de geÃ§erlidir.
```

```python
from sklearn.tree import DecisionTreeRegressor

dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, y_train)

y_pred = dt_model.predict(X_test)

print(f"Mean Squared Error (MSE): {mean_squared_error(y_test, y_pred)}")
print(f"Root Mean Squared Error (RMSE): {np.sqrt(mean_squared_error(y_test, y_pred))}")
print(f"Mean Absolute Error (MAE): {mean_absolute_error(y_test, y_pred)}")
print(f"Mean Absolute Percentage Error (MAPE): {np.mean(np.abs((y_test - y_pred) / y_test)) * 100 }")
print(f"R^2 Score: {r2_score(y_test, y_pred)}")
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': [None, 'sqrt', 'log2']
}

grid_search = GridSearchCV(DecisionTreeRegressor(random_state=42), param_grid, cv=5, scoring='r2', n_jobs=-1)
grid_search.fit(X_train, y_train)

print("En iyi parametreler:", grid_search.best_params_)
best_dt_model = grid_search.best_estimator_

y_pred = best_dt_model.predict(X_test)

print(f"Mean Squared Error (MSE): {mean_squared_error(y_test, y_pred)}")
print(f"Root Mean Squared Error (RMSE): {np.sqrt(mean_squared_error(y_test, y_pred))}")
print(f"Mean Absolute Error (MAE): {mean_absolute_error(y_test, y_pred)}")
print(f"Mean Absolute Percentage Error (MAPE): {np.mean(np.abs((y_test - y_pred) / y_test)) * 100 }")
print(f"R^2 Score: {r2_score(y_test, y_pred)}")
```
